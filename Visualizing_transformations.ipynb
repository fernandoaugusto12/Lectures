# **Visualizing Linear and Non-Linear Transformations implemented by a Neuron**

**Generate a 2D(plane) scatter plot of normally(gaussian) distributed points**
\begin{equation}
p(x,y)=\dfrac{1}{2\pi\sigma}\exp{\left(\frac{x^2+y^2}{\sigma^2}\right)}
\end{equation}

```
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib import colors
import numpy as np
import random
import math as m
#
# Random 2D gaussian distribution in a plane
#
plt.figure(figsize=(5,5),dpi=100)
plt.xlabel("X1")
plt.ylabel("X2")
npoints=5000
x = np.random.normal(size=[npoints, 2])
c=np.sqrt( x[:,0]**2 + x[:,1]**2 )
c=c*50.
plt.scatter(x[:,0],x[:,1], s=10, label='Normal Distribution', c=c, cmap = 'plasma')
plt.show()
plt.close()
#plt.savefig("name")
print(" ")
print(" ")


```

**Apply a linear transformation to the normal distribution that squezes (projects) the distribution along the X-axis**
\begin{equation}
M=
\begin{pmatrix} 
s_{11} & 0 \\
0 & s_{22} 
\end{pmatrix}
\end{equation}

```
#
# Linear Transformation of the distribution acording to T ( stretching along X and compressing along Y)
#
s11=1.0
s22=0.001
T=np.array([[s11,0],[0,s22]])
print("Matrix T=",T)
print(" ")
print(" ")
y=np.tensordot(x,T,axes=1)
plt.figure(figsize=(5,5),dpi=100)
plt.xlabel("X1")
plt.ylabel("X2")
plt.scatter(y[:,0],y[:,1], s=10, c=c, cmap = 'plasma', label='Transformed Normal Distribution')
plt.show()
plt.close()
```

**Apply a linear transformation to the normal distribution that squezes (projects) the distribution along the Y-axis**
\begin{equation}
M=
\begin{pmatrix} 
s_{11} & 0 \\
0 & s_{22} 
\end{pmatrix}
\end{equation}

```
#
# Linear Transformation of the distribution acording to T ( compressing along X and stretching along Y)
#
s11=0.001
s22=1.0
T=np.array([[s11,0],[0,s22]])
print("Matrix T=",T)
print(" ")
print(" ")
y=np.tensordot(x,T,axes=1)
plt.figure(figsize=(5,5),dpi=100)
plt.xlabel("X1")
plt.ylabel("X2")
plt.scatter(y[:,0],y[:,1], s=10, c=c, cmap = 'plasma', label='Transformed Normal Distribution')
plt.show()
plt.close()
#
```

**Apply a linear transformation to the normal distribution that squezes (projects) and rotates by an angle**
\begin{equation}
M=
\begin{pmatrix} 
cos(\phi) & -sin(\phi) \\
sin(\phi) & cos(\phi) 
\end{pmatrix}
\end{equation}

```
#
# Linear Transformation of the distribution acording to T ( compressing and rotating )
#
phi=-0.3
r11=m.cos(phi)
r12=-m.sin(phi)
r21=m.sin(phi)
r22=m.cos(phi)
T1=np.array([[r11,r12],[r21,r22]])
print("Matrix T1=",T1)
print(" ")

s11=1.0
s22=0.1
T2=np.array([[s11,0],[0,s22]])
print("Matrix T2=",T2)
print(" ")

T=np.tensordot(T2,T1,axes=1)
print("Matrix T=",T)
print(" ")

y=np.tensordot(x,T,axes=1)
print("Y=",y)
plt.figure(figsize=(5,5),dpi=100)
plt.xlabel("X1")
plt.ylabel("X2")
plt.scatter(y[:,0],y[:,1], s=10, c=c, cmap = 'plasma', label='Transformed Normal Distribution')
plt.show()
plt.close()
#
```

**Apply a NON-linear (Tanh) transformation to the normal distribution**

```
#
# Non-Linear transform the distribution ( TANH stretches the gaussian distribution to a square region )
#
y=np.tanh(x)
print("Y=",y)
plt.figure(figsize=(5,5),dpi=100)
plt.xlabel("X1")
plt.ylabel("X2")
plt.scatter(y[:,0],y[:,1], s=10, c=c, cmap = 'plasma', label='Transformed Normal Distribution')
plt.show()
plt.close()
#
```

```
#
# Non-Linear transform the distribution ( TANH stretches distribution to a square and zoom factor squezes to the edges )
#
zoom=3.5
y=np.tanh(zoom*x)
print("Y=",y)
plt.figure(figsize=(5,5),dpi=100)
plt.xlabel("X1")
plt.ylabel("X2")
plt.scatter(y[:,0],y[:,1], s=10, c=c, cmap = 'plasma', label='Transformed Normal Distribution')
plt.show()
plt.close()
#
```
