```
import random
import numpy as np
import matplotlib.pyplot as plt

R = 1.0
n_points = 100000

# Generate a an 2-D array of random points in [0,1] interval
points = np.random.rand(n_points,2)

# Put the points in a square box of side 2R , centered at the origin (0,0) 
points = 2*R*points - R

x = points[:,0]
y = points[:,1]

circle_points = points[x**2+y**2<R**2]
x_circle=circle_points[:,0]
y_circle=circle_points[:,1]


pi = 4.0 * len(circle_points[:,0]) / len(points[:,0]) 
print('Pi= ', pi)

plt.scatter(x,y,c='red')
plt.scatter(x_circle,y_circle,c='blue')

plt.show()
```
